```python
# Define a function to clean text

def clean_text(text):
    text=re.sub(r"[^a-zA-Z\s]","",text) # Keep only letters and spaces
    text=re.sub(r"\s+"," ",text).strip() # Remove extra spaces
    return text
```

Since the size of the feedback is huge, we need to divide the entire feedback in chunks of data and summarize each chunk. This summarized chunk then can be collated and formed one single summary

```python
# Creating chunks of size 8K with overlapping content of 200 characters
from langchain.text_splitter import RecursiveCharacterTextSplitter
max_chunk_size=8000
chunk_overlap=200
```

```python
text_splitter=RecursiveCharacterTextSplitter(
    chunk_size=max_chunk_size,
    chunk_overlap=chunk_overlap,
    separators=["\n\n","\n"," ",""]
)
```

```python
chunks=text_splitter.split_text(collective_feedback)
```

```python
len(chunks)
```

```python
len(chunks[23])
```

```python
#Creating array of Summary of each chunk
summaries=[]
for i,chunk in enumerate(chunks):
    response=client.chat.completions.create(
        model='gpt-4o',
        messages=[
            {"role":"system","content":'''You are a helpful assistant that summarizes feedback.Summarize the below feedback and also share the areas where the below feedback was generated. the response can be in the form of {area:"Hospitality",feedback:"The candidate got to learn a range of things..."}'''},
            {"role":"user","content":f"Summarize the following feedback:\n{chunk}"}
        ],
        max_tokens=3000
        )
    print(response)
    summary=response.choices[0].message.content
    summaries.append(summary)

final_combined_summary='\n'.join(summaries)

```
